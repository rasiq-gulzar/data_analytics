```python
import pandas as pd
students_data_one={
    'name':['John','Alice','Bob','Dylan','Eva'],
    'age':[21,22,23,24,25],
    'grade':[1,2,3,4,5],
    'subject':['Math','English','Science','History','Geography']
}
students_data_one
df1=pd.DataFrame(students_data_one)
df1.insert(4,'address',['Delhi','Mumbai','Kolkata','Chennai','Bangalore'])
df1.insert(5,'phone',[1234567890,1234567891,1234567892,1234567893,1234567894])
```

```python
student_one_roll={
    'dept':['CSE','ECE','EEE','MECH','CIVIL'],
}
df3=pd.DataFrame(student_one_roll)
df3.insert(1,'grade',df1['grade'])
df3
```

```python
students_data_two={
    'name':['vik','mavrick','mark','david','evlyn'],
    'age':[21,22,23,24,25],
    'grade':[1,2,3,4,5],
    'subject':['Math','English','Science','History','Geography']
}
df2=pd.DataFrame(students_data_two)
df2.insert(4,'address',['Delhi','Mumbai','Kolkata','Chennai','Bangalore'])
df2.insert(5,'phone',[1234567890,1234567891,1234567892,1234567893,1234567894])
```

```python
students_two_roll={
    'dept':['HR','IT','HR','IT','HR'],
}
df4=pd.DataFrame(students_two_roll)
df4.insert(1,'grade',df2['grade'])
df4
```

```python
#merge the two students inforamtion tables
df5=pd.concat([df1,df2],ignore_index=True)
df5
```

```python
student_one_with_roll=pd.merge(df1,df3,on='grade')
student_two_with_roll=pd.merge(df2,df4,on='grade')
student_one_with_roll
student_two_with_roll
```

```python
#merge the two students information with the dept table
student_big_data=pd.concat([student_one_with_roll,student_two_with_roll],ignore_index=True)
student_big_data
```
