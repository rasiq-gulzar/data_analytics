```python
df.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
df1=df.drop_duplicates('Customer Id')
df1.shape
df1['Do_Not_Contact']=df1['Do_Not_Contact'].str.capitalize()
df1['Gender']=df1['Gender'].str.capitalize()
df1['Gender']=df1['Gender'].replace({'Women':'Female','man':'Male'})
df1.drop(['SibSp','alone'],axis=1,inplace=True)

df1
```

```python
#now i have to make a space between the words on the df['Name'] using regex
df1['Name']=df1['Name'].str.replace('([A-Z])',' \\1',regex=True)
df1['Do_Not_Contact']=df1['Do_Not_Contact'].replace({'Y':'Yes','N':'No'})
```
